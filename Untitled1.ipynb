```python
import pandas as pd
data = pd.read_csv('Ranking.csv')
```

```python
# To sort universities by Overall Score in descending order
sorted_data = data.sort_values(by='OverAll Score', ascending=False)
print(sorted_data)
```

```python
data['No of student'] = data['No of student'].replace(',','', regex=True)
data['No of student per staff'] = data['No of student per staff'].replace(',','', regex=True)

data['No of student'] = data['No of student'].astype(int)
data['No of student per staff'] = data['No of student per staff'].astype(int)



data['Avg Students per Staff'] = data['No of student'] / data['No of student per staff']
print(data)
```

```python
filtered_data = data[(data['Location'] == 'United States') & (data['No of student'] > 40000)]
len(filtered_data)
```

```python
mean_students_per_staff = data['No of student per staff'].mean()
mean_students_per_staff
```

```python
missing_data = data.isna().sum()
missing_data
```

```python
data['Total Score'] = data['Teaching Score'] + data['Research Score']
data['Total Score'].head(10)
```

```python
pivoted_data = data.pivot(index='Location', columns='Name of University', values='OverAll Score')
pivoted_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['University Rank'], data['OverAll Score'])
plt.xlabel('University Rank')
plt.ylabel('Overall Score')
plt.title('University Rank vs. Overall Score')
plt.show()
```

```python
# Histogram: Distribution of Overall Score

plt.figure(figsize=(10, 6))
plt.hist(data['OverAll Score'], bins=20, edgecolor='k')
plt.xlabel('Overall Score')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Score')
plt.show()
```

```python
numeric_data = data.select_dtypes(include='number')

correlation_matrix = numeric_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
